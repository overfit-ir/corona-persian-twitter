```python
import selenium
from selenium import webdriver
import time


driver = webdriver.Firefox()
```

```python

url = 'https://twitter.com/artingarmsirii/status/1336399700375822339'
driver.get(url)

time.sleep(8)
```

```python
from bs4 import BeautifulSoup
import re
# driver.execute_script("window.scrollTo(0, 3000)") 
# html = driver.page_source
# soup = BeautifulSoup(html, "html.parser")
# d = soup.find_all(text=re.compile('Replying to')) 
# soup.find_all()

import time
pos = 0
speed = 800
replies = set()
# driver.execute_script("window.scrollTo("+str(pos)+", document.body.scrollHeight)") 
while pos < 80000:
    
    # driver.execute_script("window.scrollTo("+str(pos)+", document.body.scrollHeight)") 
    # time.sleep(.1)
    
    html = driver.page_source
    load_more_buttons = driver.find_elements_by_xpath('//*[contains(text(), "Show more replies")]')
    page_height = driver.execute_script("return document.body.scrollHeight")
    # pos = pos+speed
    
    driver.execute_script("window.scrollTo("+str(pos)+", " +str(min(pos+speed, page_height))+")") 
    # driver.execute_script("window.scrollTo("+str(pos)+", " +str(page_height)+")") 
    pos = pos + speed #page_height #(pos + page_height) / 2
    if len(load_more_buttons) > 0:
        print('load more')
        load_more_buttons[0].click()
    # print(load_more_buttons)
    soup = BeautifulSoup(html, "html.parser")
    divs = soup.find_all(text=re.compile('Replying to')) 
    for div in divs:
        # print(div.parent.parent.next_sibling.text)
        # print(div.parent.parent.parent.previous_sibling.text)
        d1 = div.parent.parent.parent.previous_sibling
        name_uname_iterator = d1.a.div.children
        name = next(name_uname_iterator).text
        username = next(name_uname_iterator).text
        datetime = d1.time['datetime']
        text = div.parent.parent.next_sibling.text
        # print(name, username, datetime, text)
        replies.add((name, username, datetime, text))
    print(len(replies), pos, page_height)
```
